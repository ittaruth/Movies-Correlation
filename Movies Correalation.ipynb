```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)

df=pd.read_csv(r"C:\Users\srava\Downloads\movies.csv")
```

```python
df.head()
```

```python
for col in df.columns:
    pct_nullvalues=np.mean(df[col].isnull())
    print("{} - {}%".format(col,pct_nullvalues))
```

```python
df=df.dropna()
```

```python
df.dtypes
```

```python
df.drop_duplicates()
```

```python
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
sns.regplot(x='budget',y='gross',data=df)
```

```python
sns.regplot(x='votes',y='gross',data=df)
```

```python
correlation_matrix=df.corr(numeric_only=True,method='pearson')
correlation_matrix
```

```python
df
```

```python
sns.heatmap(correlation_matrix,annot=True)
plt.xlabel('Features')
plt.ylabel('Features')
plt.title("Correlation Matrix for Numeric Features")
plt.show()
```

```python
df_numerized=df.copy()

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype=="object"):
        df_numerized[col_name]=df_numerized[col_name].astype("category")
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized
```

```python
tcorelation_matrix=df_numerized.corr(method="pearson")
tcorelation_matrix
```

```python
sns.heatmap(tcorelation_matrix,annot=True)
plt.title("Correlation between featues")
plt.show()
```

```python
corr_pairs=tcorelation_matrix.unstack()
print(corr_pairs)
```

```python
sorted_pairs= corr_pairs.sort_values(kind="quicksort")
print(sorted_pairs)
```

```python
strong_pairs=sorted_pairs[abs(sorted_pairs)>0.5] 
print(strong_pairs)
```
